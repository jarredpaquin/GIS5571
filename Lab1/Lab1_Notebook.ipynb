```python
import arcpy
import requests
import json
import pprint
import zipfile
```

```python
arcpy.env.workspace = r"C:\Users\jpzoo\Desktop\ArcOne\GIS5571\Lab1"
```

```python
tag = "Transit Routes"
tag2 = "Transit Stops"
base_url = "https://gisdata.mn.gov/api/3/action/package_search?q="
full_url = base_url + tag
full_url2 = base_url + tag2
```

```python
package_information = requests.get(full_url, auth = ('user', 'pass'), verify = False)
```

```python
package_dict = json.loads(package_information.content)
pprint.pprint(package_dict)
```

```python
routes_url = package_dict["result"]["results"][0]["resources"][4]["url"]
routes_url
```

```python
package_information2 = requests.get(full_url2, auth = ('user', 'pass'), verify = False)
```

```python
package_dict2 = json.loads(package_information2.content)
pprint.pprint(package_dict2)
```

```python
stops_url = package_dict2["result"]["results"][0]["resources"][0]["url"]
stops_url
```

```python
r = requests.get(routes_url)
open("shp_trans_transit_routes.zip","wb").write(r.content)
```

```python
with zipfile.ZipFile("shp_trans_transit_routes.zip", "r") as zip_ref:
    zip_ref.extractall("C:/Users/jpzoo/Desktop/ArcOne/GIS5571/Lab1")
```

```python
r2 = requests.get(stops_url)
open("shp_trans_transit_stops.zip","wb").write(r2.content)
```
